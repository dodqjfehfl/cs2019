```
def get_formatted_name(first_name,last_name,middle_name=''):
  if middle_name:
    full_name=first_name+' '+middle_name+' '+last_name
  else:
    full_name=first_name+' '+last_name
  return full_name

musician=get_formatted_name('jimi','hendrix')
print(musician)

musician=get_formatted_name('louis','armstrong','daniel')
print(musician)
```

```
def avg(*scores):
  sum=0
  for score in scores:
      sum+=score
  return sum/len(scores)

print(avg(84,69,75))
print(avg(87,90,89,95))
```

```
def grade(*scores):
  sum=0
  for score in scores:
      sum+=score
  avg=sum/len(scores)

  if avg>=90:
    return 'A'
  elif avg>=80:
    return 'B'
  elif avg>=70:
    return 'C'
  else:
    return 'F'

print(grade(84,69,75))
print(grade(87,90,89,95))
```
